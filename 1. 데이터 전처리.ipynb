## 2. 데이터 정제

### 2-1. 의미 없는 값 삭제 (describe 상 모든게 0이었음)

```python
drop_cols = {'CO2_2', '온도_2', '습도_2', '조도_2', 'CO2_3', '온도_3', '습도_3', '조도_3', 'CO2_4', '온도_4', '조도_4', '습도_4'}

df1 = df1.drop(columns=drop_cols)
df2 = df2.drop(columns=drop_cols)
df3 = df3.drop(columns=drop_cols)
df16 = df16.drop(columns=drop_cols)
```

### 2-2. 동작 및 작동은 1과 0으로 변환 (수치형)

```python
# 값이 동작 혹은 작동인 칼럼들
cols = ['CO2 릴레이동작', '난방 릴레이 동작', '냉방 릴레이 동작', '가습 릴레이 동작', '조명 릴레이 동작', '알람 릴레이 동작', '제습 릴레이 동작']
```

```python
# 변환
df1[cols] = df1[cols].replace({'동작':1, '정지':0})
df2[cols] = df2[cols].replace({'동작':1, '정지':0})
df3[cols] = df3[cols].replace({'동작':1, '정지':0})
df16[cols] = df16[cols].replace({'동작':1, '정지':0})
```
